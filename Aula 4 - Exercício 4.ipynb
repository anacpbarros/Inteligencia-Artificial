```python
#Equação da reta 

from numpy import *
import matplotlib.pyplot as plt

#Extract columns

#Variáveis
x = array ([4,6,8])
y = array ([2,3,4])

#y = ax + b -> y = mx + b
#m - coeficiente angular
#b - coeficiente linear

m = (len(x)*cov(x, y)[0,1])/((len(x) - 1)*var(x))
#m = np.cov(x, y)[0,1]/np.var(x)

b = mean(y) - m * mean(x)

print(m)
print(b)
```

```python
#Gráfico de Dispersão

#Plot dataset
plt.scatter(x, y)

#Predict y values
pred = m * x + b
#pred1 = m1 * x + b1

#Plot predictions as line of best fit
plt.plot(x, pred, c='r')
#plt.plot(x, pred1, c='r')

plt.xlabel('Variável independente')
plt.ylabel('Variável Dependente')
plt.title('Gráfico de Dispersão - Reta de Regressão')
plt.show()
```
