```python
import pandas as pd
import datetime

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bancocrai2014a2019.csv', sep=';', encoding='iso8859_2')
```

```python
df[df['data_entrada_brasil'] == 'Năo informado']
```

```python
df['data_entrada_brasil_formatado'] = \
    df[df['data_entrada_brasil'] != 'Năo informado']['data_entrada_brasil'].apply(
        lambda x: datetime.datetime.strptime(x, '%d/%m/%Y')
    )
df.head()
```

```python
#cria 'data_cadastro_formatado' convertendo os dados de 'data_cadastro' para datetime
df['data_cadastro_formatado'] = pd.to_datetime(df['data_cadastro'])

#cria 'ano_cadastro' e 'ano_entrada' filtrando apenas os anos das duas variáveis
df['ano_cadastro'] = df['data_cadastro_formatado'].apply(lambda x: (x.year))
df['ano_entrada'] = df['data_entrada_brasil_formatado'].apply(lambda x: (x.year))

#converte a variável 'ano_entrada' para int, a fim de remover a casa decimal após o ano
df['ano_entrada'] = df['ano_entrada'].fillna(0)
df['ano_entrada'] = df['ano_entrada'].astype(int)

#cria 'diferença_anos' e calcula a diferença de anos entre a entrada no brasil e a data do cadastro em anos
df['diferença_anos'] = df['ano_cadastro'] - df['ano_entrada']

df.head()
```
